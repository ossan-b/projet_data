## Conso pour chaque pays

```python
import numpy as np
world_alcohol = np.genfromtxt('world_alcohol.csv', delimiter=',', dtype='U75', skip_header=1)

totals = {}
#selection des lignes correspondantes à 1989
year = world_alcohol[world_alcohol[:, 0] == '1989']
year
```

```python
# liste de tous les pays
countries = world_alcohol[:, 2]
countries
```

```python
# parcourir la liste des pays et pour chaque pays, selection les lignes de year qui correspondent à ce pays

for country in countries:
    is_country = year[:, 2]== country
    country_consumption = year[is_country, :]
    alcohol_column = country_consumption[:, 4]
    is_empty = alcohol_column == ''
    alcohol_column[is_empty]= '0'
    alcohol_column = alcohol_column.astype(float)
    totals[country] = alcohol_column.sum()
    
```

```python
print(totals)
```

```python
totals['Papua New Guinea']
```

### le pays qui a le plus conso en 1989

```python
highest_value = 0
highest_key = None
for i in totals:
    if totals[i]> highest_value:
        highest_value = totals[i]
        highest_key = i
```

```python
def calculate_major_cat_totals(df):
    cat_unique = df['Major_category'].unique()
    counts_dictionary = {}
    
    for c in cat_unique:
        major_df = df[df['Major_category'] == c]
        total = df['Total'].sum()
        counts_dictionary[c] = total
    return counts_dictionary

    
```

```python
# écrire une fonction qui convertir une chaine de caractères en un entien
def extract_age(age_str):
    if pd.isnull(age_str):
        return None
    age_str = age_str.split(' ')[0]
    age_str = age_str.replace('+', ' ')
    return int(age_str)

#une nouvelle col 'int_age'
data['int_age']= data['Age'].apply(extract_age)
data['int_age'].head()
```

### Convertir les revenus en valeurs numeriques

```python
data['How much total combined money did all members of your HOUSEHOLD earn last year?'].value_counts()
```

```python
def extract_money(money_str):
    if pd.isnull(money_str):
        return None
    #séparateur de chaine de caractere ' ', reccuperer chaque fois le 1ier élément de la liste
    money_str = money_str.split(' ')[0]
    if money_str == 'Prefer':
        return None
    money_str = money_str.replace('$', '')
    money_str = money_str.replace(',', '')
    
    return int(money_str)

```

```python
data['int_money'] = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(extract_money)
```

```python
import matplotlib.pyplot as plt
from numpy import arange
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

#largeur des barre
#position des barres
bar_widths = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 1
tick_positions = arange(1, 6)
fig, ax = plt.subplots()

ax.barh(bar_positions, bar_widths, 0.5)

ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_ylabel('Sources de notation')
ax.set_xlabel('Note moyenne')
ax.set_title('Moyenne des notes utilisateurs pour le film Aventagers: Age of Ultron (2015)')

plt.show()

```

### Comparaison d'histogrammes

#### Tracer 4 graphiques

```python
import pandas as pd
revviews = pd.read_csv('fandango_scores.csv')
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']

norm_reviews = revviews[cols]
```

```python
fig = plt.figure(figsize=(5, 20))
ax1 = fig.add_subplot(4, 1, 1)
ax2 = fig.add_subplot(4, 1, 2)
ax3 = fig.add_subplot(4, 1, 3)
ax4 = fig.add_subplot(4, 1, 4)

# faire varier le nombre de classes  à 20
ax1.hist(norm_reviews['Fandango_Ratingvalue'], bins = 20, range=(0, 5))
ax1.set_title('Distribution des notes de Fandango')
ax1.set_ylim(0, 50)


ax2.hist(norm_reviews['RT_user_norm'], bins = 20, range=(0, 5))
ax2.set_title('Distribution des notes de Rotten Tomatoes')
ax2.set_ylim(0, 50)



ax3.hist(norm_reviews['Metacritic_user_nom'], bins = 20, range=(0, 5))
ax3.set_title('Distribution des notes de Metacritic')
ax3.set_ylim(0, 50)


ax4.hist(norm_reviews['IMDB_norm'], bins = 20, range=(0, 5))
ax4.set_title('Distribution des notes de IMDB')
ax4.set_ylim(0, 50)

plt.show()
```
